### 2. Assemble a list of non ignition points

Here we create a dataset with the same size as the ignition points dataset. A random location & date is picked.

```python
grid_data = gpd.read_file(path + 'grid/europe_10km_grid.shp') 
```

```python
grid_data['id'] = grid_data.id.astype('int32')
```

```python
grid_data
```

```python
for remove_id in ids_to_remove:
    grid_data = grid_data[grid_data.id != remove_id]
```

```python
grid_data = grid_data.reset_index(drop = True)
```

```python
location_ids = grid_data.id.unique() # all locations
dates = pd.date_range('2008-01-07', '2023-12-24') # all dates apart from first 7 and last 7 due to rolling means and sums
dates = dates[ ~((dates.month == 2) & (dates.day == 29))] # remove leap year days 

for m in months:
    dates = dates[ ~(dates.month == m)] 
```

#### Select random ids

```python
t0 = time.time()
df_non_ignit = pd.DataFrame()
df_non_ignit['id'] = ''
df_non_ignit['date'] = ''

i = 0

while len(df_non_ignit) != len(df_ignit):

    choice_id = np.random.choice(location_ids) # select a random id
    choice_date = np.random.choice(dates) # select a random date

    # Make sure to not select the same combination of id /
    # and date as in the df_ignit 
    if ((choice_id == df_ignit.id) & (choice_date ==  df_ignit.date.astype('string'))).any():
        print('Historical ignition location / date')
        continue
    elif ((choice_id == df_non_ignit.id) & (choice_date ==  df_non_ignit.date.astype('string'))).any():
        print('Location / date combination already in list')
        continue
    else:     
        df_non_ignit.loc[i, ['id']] = choice_id
        df_non_ignit.loc[i, ['date']] = choice_date
        i = i + 1
t1 = time.time()
print(t1 -t0)
```

```python
df_non_ignit['date'] = df_non_ignit['date'].astype('datetime64[ns]')
df_non_ignit['ignit'] = 0
```

```python
df_all = pd.concat([df_ignit, df_non_ignit], ignore_index = True)
```

```python
df_all['date'] = df_all.date.astype('datetime64[ns]')
```

```python
df_all = df_all.set_index('id')
```
